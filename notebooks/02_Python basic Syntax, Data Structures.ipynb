Handling elements

```python

s = {1, 2, 3}

s.add(4)        # Add element
s.remove(2)     # Remove element (error if not found)
s.discard(10)   # Safe remove (no error)
s.pop()         # Remove random element
```

Membership Check

```python
3 in s   # True
10 in s  # False
```

Set Algebra

```python
segment_a = {1, 2, 3}
segment_b = {3, 4, 5}
```

Intersection (common elements)

```python
segment_a.intersection(segment_b)
# or
segment_a & segment_b
```

Union (all unique elements)

```python
segment_a.union(segment_b)
# or
segment_a | segment_b
```

Difference

```python
segment_a.difference(segment_b)
# or
segment_a - segment_b
```

Symmetric Difference

```python
segment_a.symmetric_difference(segment_b)
# or
segment_a ^ segment_b
```
